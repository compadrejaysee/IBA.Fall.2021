```python
import numpy as np
from random import random
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
```

Defining class for Neural network with hidden layer and sigmoid activation function. It will contain 4 neurons at output layer. weight matrices will be initialized using random normal function provided by numpy.

```python
class NN_1_sigmoid():
    def __init__(self,input_size,hidden_size):
        self.hidden_size=hidden_size
        self.input_size=input_size
        self.output_size=4
    #input_size=int(input('Enter the number of features'))
    #hidden_size=int(input('Enter number of neurons in the hidden layer'))
    #output_size=int(input('Enter number of neurons in the output layer'))
        self.w1=np.random.normal(scale=1,size=(self.input_size,self.hidden_size))
        self.w2=np.random.normal(scale=1,size=(self.hidden_size,self.output_size))
       # y = pd.get_dummies(y).values
        #N = y.size
        
    def sigmoid(self,x):
        return 1 / (1 + np.exp(-x))
    def tanh(self,Z):
        a=(np.exp(Z)-np.exp(-Z))/(np.exp(Z)+np.exp(-Z))   
        return a
    def loss_function(self,predicted, true):
        return ((predicted - true)**2).sum() / (2*predicted.size)#using mean squared error as the loss function
    def forward(self,X,y):
        z1=np.dot(X,self.w1)
        a1=self.sigmoid(z1)
        z2=np.dot(a1,self.w2)
        a2=self.sigmoid(z2)
        cost=self.loss_function(a2,y)
        return cost,a2,a1
    def back(self,cost,learning_rate,a2,a1,N):
        
        e1 = a2-y
        dw1 = e1 * a2* (1 - a2)
        e2 = np.dot(dw1, self.w2.T)
        dw2 = e2 * a1 * (1 - a1)
        # updating weights on the basis of gradients
        deltaw2 = np.dot(a1.T, dw1) / N
        deltaw1 = np.dot(X.T, dw2) / N

        self.w2 = self.w2 - learning_rate * deltaw2
        self.w1 = self.w1 - learning_rate * deltaw1
    def train(self,X,y,learning_rate,epochs):
        #y = pd.get_dummies(y).values
        N = y.size
        results = pd.DataFrame(columns=["mse"])
        #w1=np.random.normal(scale=1,size=(self.input_size,hidden_size))
        #w2=np.random.normal(scale=1,size=(self.hidden_size,self.output_size))
        for i in range(epochs):
            cost,a2,a1=self.forward(X,y)
            self.back(cost,learning_rate,a2,a1,N)
            results=results.append({"mse":cost},ignore_index=True )
        m=results.mse.plot(title="Mean Squared Error")
        plt.show()
    def predict(self,X):
        z1=np.dot(X,self.w1)
        a1=self.sigmoid(z1)
        z2=np.dot(a1,self.w2)
        a2=self.sigmoid(z2)
        return a2
   # learning_rate=float(input('Enter learning rate'))
    #epochs=int(input('Enter number of epochs'))
    #train(X,y,learning_rate,epochs,input_size,hidden_size,output_size)
    
   # m=results.mse.plot(title="Mean Squared Error")
    
   # plt.show()
    
```

Demonstration of neural network with sigmoid activation and 1 hidden layer

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=15000,n_features=4,n_classes=4,n_informative=4,n_redundant=0)
actual_y=y
y=pd.get_dummies(y).values
```

```python
nn=NN_1_sigmoid(4,2)#First argument takes number of inputs and second argument takes number of neurons in hidden layer

nn.train(X,y,0.5,1000)#learning rate set as 0.5. Number of epochs=1000
```

```python
y_pred=nn.predict(X)
```

Using 4 neurons at the output layer outputs the probability of each class label at each neuron. For example neuron zero will output the probability of the class being zero. Using the np.argmax function class labels will be assigned. 

```python
predictions=pd.DataFrame(y_pred)

#for i in range(len(predictions)):
#    m=predictions.loc[i].max()
#    indx=predictions.loc[i].loc[predictions.loc[i]==m].index[0]
#    predictions.loc[i,'label']=indx
predictions['label']=np.argmax(y_pred,axis=1)

#predictions

predictions['actual_y']=actual_y

display(predictions)

from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
accuracy=accuracy_score(predictions['actual_y'], predictions['label'])*100
print('Accuracy',accuracy)
recall=recall_score(predictions['actual_y'], predictions['label'],average='macro')*100
print('Recall',recall)
f1=f1_score(predictions['actual_y'], predictions['label'],average='macro')*100
print('f1 score',f1)
print(classification_report(predictions['actual_y'], predictions['label']))
```
