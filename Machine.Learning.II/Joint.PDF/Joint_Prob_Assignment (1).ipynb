### User-Defined Functions

```python
# -------------------Discreet----------------------------------

# Takes RVs as input from the user
def input_():

  RV1 = []
  RV2 = []

  n1 = int(input("\n\nSpecify the number of values for Discreet RV1 : "))
  for i in range(0, n1):
      ele = int(input("\n\nPopulate the RV1 with values: "))
      RV1.append(ele)    
  print("\nThe values for Random Variable 1 are:", RV1)

  n2 = int(input("\n\nSpecify the number of values for Discreet RV2 : "))
  for i in range(0, n2):
      ele = int(input("\n\nPopulate the RV2 with values: "))
      RV2.append(ele)      
  print("\nThe values for Random Variable 2 are:", RV2)

  return (RV1,RV2)

# generating probabilities

def gen_prob (RV1, RV2):
  l1 = len(RV1)
  l2 = len(RV2)
  l = l1 * l2
  x = np.linspace(1,5,l)
  prob = [float(i)/sum(x) for i in x]
  return prob

# Visualize the joint prob dist of RV1 and RV2

def view_joint_prob(RV1, RV2):

  joint_dist = Table().domain('X', RV1, 'Y', RV2).probabilities(gen_prob(RV1,RV2))
  return joint_dist

# Marginal X and Y

def marginal (joint_dist):

  print_mar_x = joint_dist.marginal("X")
  mar_x = joint_dist.marginal_dist("X")
  print_mar_y = joint_dist.marginal("Y")
  mar_y = joint_dist.marginal_dist("Y")
  return mar_x, print_mar_x, mar_y, print_mar_y

# expected value X,Y

def expected_val (mar_x,mar_y):
  val_x = mar_x['Value']
  prob_x = mar_x['Probability']
  val_y = mar_y['Value']
  prob_y = mar_y['Probability']
  e_x = val_x * prob_x
  e_y = val_y * prob_y
  return (sum(e_x), sum(e_y))


# conditoinal

def conditional (RV1, RV2, joint_dist): 

  x_conditional_mean_list = []
  y_conditional_mean_list = []
  RV1.sort
  RV2.sort(reverse = True)

  cond_x = joint_dist.conditional_dist("X")
  cond_y = joint_dist.conditional_dist("Y")

  # Means

  for i in range(len(cond_x.index)-1):
    x_dist_cond = cond_x.iloc[i, :len(cond_x.columns)-1].values
    conditional_mean = x_dist_cond * RV1
    x_conditional_mean_list.append(sum(conditional_mean))
  x_conditional_mean_list

  for i in range(len(cond_y.columns)-1):
    y_dist_cond = cond_y.iloc[:len(cond_y.index)-1,i].values
    conditional_mean = y_dist_cond * RV2
    y_conditional_mean_list.append(sum(conditional_mean))
  y_conditional_mean_list

  # Variance

  x_conditional_var_list = []
  y_conditional_var_list = []

  for i in range(len(cond_x.index)-1):
    var = (RV1-x_conditional_mean_list[i])**2 * cond_x.iloc[i, :len(cond_x.columns)-1].values
    x_conditional_var_list.append(sum(var))
  x_conditional_var_list

  for i in range(len(cond_y.columns)-1):
    var = (RV2-y_conditional_mean_list[i])**2 * cond_y.iloc[:len(cond_y.index)-1,i].values
    y_conditional_var_list.append(sum(var))
  y_conditional_var_list

  conditions_x = np.array(cond_x.index)[:-1]
  conditional_x = pd.DataFrame()
  conditional_x['Conditions'] = conditions_x
  conditional_x['Conditional Mean'] = np.array(x_conditional_mean_list)
  conditional_x['Conditional Variance'] = np.array(x_conditional_var_list)

  conditional_y = pd.DataFrame()
  conditions_y = np.array(cond_y.columns)[:-1]
  conditional_y['Conditions'] = conditions_y
  conditional_y['Conditional Mean'] = np.array(y_conditional_mean_list)
  conditional_y['Conditional Variance'] = np.array(y_conditional_var_list)

  return cond_x, cond_y, conditional_x, conditional_y

#-------------------------------- Continous------------------------------------------------------

# Input continous joint function

def cont_joint_prob():
  expr = input("\nEnter the Continous Joint Probability function F(X,Y): \n")
  expr = sympify(expr)
  x_upper = input("\nEnter the Upper limit of X: \n")
  x_lower = input("\nEnter the Lower limit of X: \n")
  x_lim = [x_lower, x_upper]
  y_upper = input("\nEnter the Upper limit of Y: \n")
  y_lower = input("\nEnter the Lower limit of Y: \n")
  y_lim = [x_lower, x_upper]

  return expr, x_lim ,y_lim

# marginal

def continous_mar (expr, x_lim, y_lim):
  init_printing(use_unicode=False, wrap_line=False)
  x = Symbol('x')
  y = Symbol('y')
  cont_mar_x = integrate(expr, (y, y_lim[0], y_lim[1]))
  cont_mar_y = integrate(expr, (x, x_lim[0], x_lim[1]))
  return cont_mar_x, cont_mar_y

# expected value 

def cont_expected_val (cont_mar_x, cont_mar_y, x_lim,y_lim):

  init_printing(use_unicode=False, wrap_line=False)
  x = Symbol('x')
  y = Symbol('y')
  cont_e_x = integrate(x*(cont_mar_x), (x, x_lim[0], x_lim[1]))
  cont_e_y = integrate(y*(cont_mar_y), (y, y_lim[0], y_lim[1]))
  return cont_e_x, cont_e_y

# conditional
def cont_conditional (expr, x_lim, y_lim, cont_mar_x, cont_mar_y):

  init_printing(use_unicode=False, wrap_line=False)
  x = Symbol('x')
  y = Symbol('y')

  cont_cond_x = expr/cont_mar_y
  cont_cond_y = expr/cont_mar_x

  cont_cond_e_x = integrate(x*(cont_cond_x), (x, x_lim[0], x_lim[1]))
  cont_cond_e_y = integrate(y*(cont_cond_y), (y, y_lim[0], y_lim[1]))

  return cont_cond_x, cont_cond_y,cont_cond_e_x, cont_cond_e_y
```
