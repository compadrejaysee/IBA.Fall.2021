#Creating a standard algorithm incorporating different types of clustering

*   APC = Affinity Propagation Clustering 
*   AC = Agglomerative Clustering
*   BC = BIRCH Clustering
*   DBSC = DBSCAN Clustering
*   KMC = K-means Clustering 
*   MBKMC = Mini-Batch K-means Clustering
*   MSC = Mean Shift Clustering
*   OC = Optics Clustering 
*   SC = Spectral Clustering
*   GMC = Gaussian Mixture Clustering

```python
def clustering(algo_abbr, algo_name, X, y):
	yhat = None
	algo = algo_abbr
	if algo == 'APC':
		# define the model
		model = AffinityPropagation(damping=0.9)
		# fit the model
		model.fit(X)
		# assign a cluster to each example
		yhat = model.predict(X)
	elif algo == 'AC':
		model = AgglomerativeClustering(n_clusters=2)
		# fit model and predict clusters
		yhat = model.fit_predict(X)
	elif algo == 'BC':
		# define the model
		model = Birch(threshold=0.01, n_clusters=2)
		# fit the model
		model.fit(X)
		# assign a cluster to each example
		yhat = model.predict(X)
	elif algo == 'KMC':
		# define the model
		model = KMeans(n_clusters=2)
		# fit the model
		model.fit(X)
		# assign a cluster to each example
		yhat = model.predict(X)
	elif algo == 'MBKMC':
		# define the model
		model = MiniBatchKMeans(n_clusters=2)
		# fit the model
		model.fit(X)
		# assign a cluster to each example
		yhat = model.predict(X)
	elif algo == 'GMC':
		# define the model
		model = GaussianMixture(n_components=2)
		# fit the model
		model.fit(X)
		# assign a cluster to each example
		yhat = model.predict(X)
	elif algo == 'DBSC':
		# define the model
		model = DBSCAN(eps=0.30, min_samples=9)
		# fit model and predict clusters
		yhat = model.fit_predict(X)
	elif algo == 'MSC':
		# define the model
		model = MeanShift()
		# fit model and predict clusters
		yhat = model.fit_predict(X)
	elif algo == 'OC':
		# define the model
		model = OPTICS(eps=0.8, min_samples=10)
		# fit model and predict clusters
		yhat = model.fit_predict(X)
	elif algo == 'SC':
		# define the model
		model = SpectralClustering(n_clusters=2)
		# fit model and predict clusters
		yhat = model.fit_predict(X)

	if yhat is None:
		raise Exception("Invalid Algorithm.")

	print('\n\n----------------------------------------------------')
	print('Plotting algorithm :: ' + algo_name)
	print('----------------------------------------------------\n\n')

	for class_value in range(2):
		# get row indexes for samples with this class
		row_ix = where(y == class_value)
		# create scatter of these samples
		pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
	# show the plot
	pyplot.title("Synthetic Classification Dataset")
	pyplot.show()
 
	# retrieve unique clusters
	clusters = unique(yhat)
	# create scatter plot for samples from each cluster
	for cluster in clusters:
		# get row indexes for samples with this cluster
		row_ix = where(yhat == cluster)
		# create scatter of these samples
		pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
	# show the plot
	pyplot.title(algo_name)
	pyplot.show()
```

#Running the algorithm on a synthetic dataset

```python
algorithms = {
  "APC": "Affinity Propagation Clustering",
  "AC": "Agglomerative Clustering",
  "BC": "BIRCH Clustering",
  "DBSC": "DBSCAN Clustering",
  "KMC": "K-means Clustering",
  "MBKMC": "Mini-Batch K-means Clustering",
  "MSC": "Mean Shift Clustering",
  "OC": "Optics Clustering",
  "SC": "Spectral Clustering",
  "GMC": "Gaussian Mixture Clustering"
}
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)

for algo_abbr, algo_name in algorithms.items():
	clustering(algo_abbr, algo_name, X, y)
```

#Commentary on algorithms:

General: In general, different clustering algorithms work better in different types of datasets. We have used a mix of parameters for different kinds of dataset and the results are analyzed below of each dataset.

**Dataset 1**:
The parameters for these algorithms were mostly set to 2 clusters, and we tried to reduce most parameters like minimum samples, epsilon, damping, n_components etc. for these algorithms. 

The dataset had 17 columns, and we restricted the rows to 1000 (as the program kept crashing)


For the first dataset, Affinity Propagation and Spectral Clustering do not seem to be working at all. 

Agglomerative, Birch, DBSCAN, K-means and Gaussian Mixture do not seem to be working too well either.  

Mini-Batch K-means, Mean Shift and Optics clustering seem to be doing a better job as compared to the rest.  

Mini-batch K-means and Mean Shift are both partitioning methods, and Optics is a hierarchy based method. 

Clearly, on this type of dataset, customers can be segregared into mostly 2 types based on whether they will end up making an online purchase or not, and better results are obtained using partitioning methods.

**Dataset 2**:
The parameters for these algorithms were mostly set to 4 clusters, and we tried to increase most parameters like minimum samples, epsilon, damping, n_components etc. for these algorithms. 

The dataset had 27 columns, 774 rows.

For the second dataset, Spectral Clustering seems to be working poorly. 

Agglomerative, Affinity Propagation, Birch, DBSCAN, Mean Shift and Optics clustering seem to be giving mixed results.

K-means, Gaussian Mixture and Mini-Batch K-means seem to be doing a much better job as compared to the rest, which are again, partitioning methods.  

Clearly, on this type of dataset, for audit purposes, the customers need to be divided based on their risk levels of high, medium, low etc., and can be segregared into mostly 3 or more types, and better results are obtained using partitioning methods.

**Dataset 3:**
The parameters for these algorithms were mostly set to 2 clusters, but we tried to increase most parameters like minimum samples, epsilon, damping, n_components etc. for these algorithms. 

The dataset had 4 columns, ~200 rows.

For the third dataset, DBSCAN, Mean Shift and Spectral Clustering seem to not be working at all. 

Affinity Propagation and Optics clustering seem to be giving mixed results.

Agglomerative, Birch, K-means, Gaussian Mixture and Mini-Batch K-means seem to be doing a much better job as compared to the rest. 

These are a mix of partitioning and hierarchal methods.

On the customer dataset dataset, there were only 4 columns that included information on customers. The goal was to divide mall going customers based on their age, gender, income and genre, and since the assumption is that there should not be major disparity, the customers can mostly be divided into 2 types. This explains why the partitioning methods responded well, but the some hierarchical methods also responded well, because there were very few columns, and dividing based on hierarchy was possible. 

**Dataset 4:**
The parameters for these algorithms were mostly set to 3 clusters, and we tried to decrease most parameters like minimum samples, epsilon, damping, n_components etc. for these algorithms. 

The dataset had 113 columns, ~20,000 rows, which we restricted to 1000 because the program kept crashing.

For the fourth dataset, Agglomerative, Affinity Propagation, BIRCH, DBSCAN and Spectral Clustering seem to not be working very well.  

Mean Shift and Optics clustering seem to be giving mixed results.

K-means, Gaussian Mixture and Mini-Batch K-means seem to be doing a much better job as compared to the rest, which are again, partitioning methods.  

This is similar to the audit dataset (datset 2), whereby customers needed to be segregated based on their risk levels of high, medium, low etc., and can be segregared into mostly 3 or more types, and better results are obtained in both cases using partitioning methods.
