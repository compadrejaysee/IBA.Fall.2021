### Confidence Interval

#### Q1


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from scipy.stats import t, sem
from numpy.random import normal
mean=71
sd=2.8
n=49
sample=normal(mean,sd,n)
print('95% CI for 49 random samples: ')
print(t.interval(0.95, len(sample)-1, np.mean(sample),sem(sample)))
print(f'Error Bound={(t.interval(0.95, len(sample)-1, np.mean(sample),sem(sample))[1]-t.interval(0.95, len(sample)-1, np.mean(sample),sem(sample))[0])/2}')
plt.hist(sample,30, density=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from scipy.stats import t, sem
from numpy.random import normal
mean=71
sd=2.8
n=2000
sample=normal(mean,sd,n)
print('95% CI for 2000 random samples: ')
print(t.interval(0.95, len(sample)-1, np.mean(sample),sem(sample)))
print(f'Error Bound={(t.interval(0.95, len(sample)-1, np.mean(sample),sem(sample))[1]-t.interval(0.95, len(sample)-1, np.mean(sample),sem(sample))[0])/2}')
plt.hist(sample,30, density=1)
```

The confidence interval and error bound have decreased significantly as the number of samples have increased. SInce there are more samples therefore a smaller interval will be required to capture the true mean.

### Central Limit Theorem

```python
numIterations = np.asarray([1,2,5,10,50,100]); #number of i.i.d RVs  

experiment = 'sugarcane' 

maxNumForExperiment = {'sugarcane':12} #max height of sugarcane is 12feet

nSamp=1000000

k = maxNumForExperiment[experiment] 

fig, fig_axes = plt.subplots(ncols=3, nrows=2, constrained_layout=True) 

for i,N in enumerate(numIterations):  
    y = np.random.randint(low=4,high=k+1,size=(N,nSamp)).sum(axis=0)  #Minimum height is 4feet
    row = i//3;col=i%3;  
    bins=np.arange(start=min(y),stop=max(y)+2,step=1)  
    fig_axes[row,col].hist(y,bins=bins,density=True)  
    fig_axes[row,col].set_title('N={} {}'.format(N,experiment)) 
    
plt.show() 
```

### Probability Distributions

#### Logistic DIstribution

```python
from scipy.stats import logistic
numargs = logistic.numargs 
a, b = 4.32, 3.18
rv = logistic(a, b) 
    
print ("RV : \n", rv) 
quantile = np.arange (0.03, 2, 0.21) 
  
# Random Variates 
R = logistic.rvs(a, b) 
print ("Random Variates : \n", R) 
  
# PDF 
R = logistic.pdf(a, b, quantile) 
print ("\nProbability Distribution : \n", R) 
distribution = np.linspace(0, np.minimum(rv.dist.b, 3)) 
print("Distribution : \n", distribution) 
     
plot = plt.plot(distribution, rv.pdf(distribution)) 
x = np.linspace(-5, 6, 100) 
     
# Varying positional arguments 
y1 = logistic .pdf(x, 1, 3) 
y2 = logistic .pdf(x, 1, 4) 
plt.plot(x, y1, "*", x, y2, "r--") 
```

#### Logit-Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import logit
from scipy.stats import norm, rv_continuous


class LogitNormal(rv_continuous):
    def __init__(self, scale=1, loc=0):
        super().__init__(self)
        self.scale = scale
        self.loc = loc

    def _pdf(self, x):
        return norm.pdf(logit(x), loc=self.loc, scale=self.scale)/(x*(1-x))


fig, ax = plt.subplots()
values = np.linspace(10e-10, 1-10e-10, 1000)
sigma, mu = 1.78, 0
ax.plot(
    values, LogitNormal(scale=sigma, loc=mu).pdf(values), label='subclassed'
)
ax.legend()
fig.show()
```

The graoh shows the pdf of the logit-normal distribution. In theory, the logit-normal distribution is one in which the logit of the distribution has a normal distribution 

#### Logit Function

```python
from scipy.special import logit, expit

print(logit([0, 0.25, 0.5, 0.75, 1]))


#expit is the inverse of `logit`

expit(logit([0.1, 0.75, 0.999]))


#Plot logit(x) for x in [0, 1]:

import matplotlib.pyplot as plt
x = np.linspace(0, 1, 501)
y = logit(x)
plt.plot(x, y)
plt.grid()
plt.ylim(-6, 6)
plt.xlabel('x')
plt.title('logit(x)')
plt.show()
```

Logit function is the inverse function of the logistic distribution where by given a probability it will output the value of the random variable. This is evident from the graph
