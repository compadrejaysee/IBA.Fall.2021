## Q2: Assume that temperatures are distributed according to the standard normal distribution. Find the probability that a random temp measurement complies with the following requirements:

a)•	less than −5.5∘
b)•	less than 4.44
c)•	greater than 29.99
d)•	greater than −15.5
e)•	between 15∘ and 40∘
f)•	between −23.00∘ and −2.00∘
g)•	greater than 0∘

```python
#The standard normal distribution is a normal distribution with a mean of 0 and standard deviation of 1
mu = 0
sigma = 1

#For part (a)
a=stats.norm.cdf(-5.5,mu,sigma)
print("Probability if Random Temp is less than -5.5: ",a , " OR ", a*100,"%\n" )

#For part (b)
b=stats.norm.cdf(4.44,mu,sigma)
print("Probability if Random Temp is less than 4.44: ",a , " OR ", b*100,"%\n" )

#For part (c)
#Here we can use cdf and subtract it from 1 or we can use 'sf' i.e. Survival Function which does the same thing for us 
c = 1-stats.norm.cdf(29.99)
print("Probability if Random Temp is greater than 29.99: ",c , " OR ", c*100,"%\n" )

#For part (d)
#Here we can use cdf and subtract it from 1 or we can use 'sf' i.e. Survival Function which does the same thing for us 
d = 1-stats.norm.cdf(-15.5,mu,sigma)
print("Probability if Random Temp is greater than -15.5: ",d , " OR ", d*100,"%\n" )

#For part (e)
e = stats.norm.cdf(40,mu,sigma) - stats.norm.cdf(15,mu,sigma)
print("Probability if Random Temp is between 15 and 40: ",e , " OR ", e*100,"%\n" )

f = stats.norm.cdf(-2,mu,sigma) - stats.norm.cdf(-23,mu,sigma)
print("Probability if Random Temp is between -23 and -2: ",f , " OR ", f*100,"%\n" )

#For part (g)
#Here we can use cdf and subtract it from 1 or we can use 'sf' i.e. Survival Function which does the same thing for us 
g = stats.norm.sf(0,mu,sigma)
print("Probability if Random Temp is greater than 0: ",g , " OR ", g*100,"%\n" )
```

## Q7: The average salary for freshmen professors is PKR 50,000. Assume normal distribution with standard deviation 2140. 

•	What is the probability that a randomly selected first-year teacher makes between 15,000 and 35,000 each year?


•	What is the probability that a randomly selected freshman prof has a salary less than 22,000?

•	What is the probability that a randomly selected freshman prof has a salary at least 12,000?

•	What is 90th percentile salary?


```python


mu = 50000
sigma = 2140

#Just to have a better understanding that how closely variables are lying towards mean, I have plotted normal distribution graph.
plot_x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(plot_x, stats.norm.pdf(plot_x, mu, sigma))
plt.show()

# --------------------------------------------------------------------------------
# For part(a)

# NOTE: p should less than 35000(a) and greater than 15000(b)
# p<a and p>b

a_z_score = (35000-mu)/sigma
b_z_score = (15000-mu)/sigma

a = round(stats.norm.cdf((a_z_score)),3)
b = round(stats.norm.cdf((b_z_score)),3)
# print(a)
# print(b)
part_a= a-b
print("Part A => The probability between 15000 and 35000 is: ",part_a)

# --------------------------------------------------------------------------------
# For part(b)
# p<20000

z_score = (22000-mu)/sigma
part_b = round(stats.norm.cdf((z_score)),3)
print("Part B => The probability less than 22000 is: ",part_b)

# --------------------------------------------------------------------------------
# For part(c)
z_score_12 = (12000-mu)/sigma
part_c = round(stats.norm.sf((z_score_12)),3)
print("Part C => The probability atleast 12000 is: ",part_c)


# --------------------------------------------------------------------------------
# For part(d)
z = (stats.norm.ppf(0.90))
x = (z*sigma)+50000
print("Part D => The probability at 90th percentile: ", round(x,2))
```

# **Confidence Interval**

## Q1: Construct a 95% confidence interval for the mean height of male Pakistanis. Forty-nine male Pakistanis are surveyed. The sample mean is 71 inches. The sample standard deviation is 2.8 inches

```python
# For part(a): State the CI interval
mu = 71
sigma = 2.8
total_len = 49
alpha=0.95

c_i = stats.norm.interval(alpha, mu, sigma/np.sqrt(total_len)) 
print("Confidence Interval: ",c_i)
```

```python
## For part(b):	Draw the graph
import seaborn as sns
  
#Finding z-critical to find max and min value
z_crit = stats.norm.ppf(1-0.05/2)
print("Z-Critical/z: ", z_crit)

val = z_crit*sigma
max_val = mu + val
min_val = mu - val
print("min_val: ",min_val," max_val: ",max_val)

# generate random data
np.random.seed(0)
x = np.random.randint(min_val, max_val, total_len)
y = x+np.random.normal(mu, sigma, total_len)
  
# create lineplot
ax = sns.lineplot(x, y, ci=95)
```

```python
# For part(c):	Calculate the error bound
err_bound = c_i[1] - mu
print("Error bound: ",err_bound)
```

##d) What will happen to the level of confidence obtained if 2,000 male Pakistanis are surveyed instead of 49? Why?

If we increase the size of sample n, the level of confidence would decrease. 
Because it would be taking less area under the normal curve (which translates into a smaller interval) to capture the true population mean.

## Q3: The average height of children has a normal distribution with standard deviation of 3.4 inches. We are interested in estimating the mean height of children at school to within two inches with 93% confidence. How many children must you measure?

```python
sigma = 3.4
err_bound = 2
alpha = 1-0.93 #93% given
z = stats.norm.ppf(1-alpha/2)
# print(z)
n = ((z**2)*(sigma**2))/(err_bound**2)
print("Answer: ",n.round()," children should be measured in order to be 93% confident.")
```

# **Central Limit Theorem**
**Q1: Model any problem from the agricultural domain in Python and demonstrate the central limit theorem (like dice roll and coin toss example).**

**Answer:**
Lets suppose there are n number of farms spread on x acres. The aggregation of crop yield distribution varies from acre to acre and farm to farm.
The crop yield from each acre is not normally distributed as the type of crop, weather and cultivation method varies. 
However, if we examine the applicability of the central limit theorem (CLT) to aggregate crop harvested we might be able to shift the non-normal distribution to normal distribution.
If we take average of each acre’s annual yield irrespective of conditions effecting on yield. We will be able to see that from the averages taken from numerous acres, the average of each acre is closer to the mean of all acres.

Thus, we can say that by applying CLT, we have somehow found some normality in our data


```python
import numpy
import matplotlib.pyplot as plt
 
# number of acres we are taking. Increasing from left to right
num = [10, 50, 150, 250] 
# list of average of yield we got each acres
means = [] 
 
# Generating 1, 10, 30, 100 random numbers from 100 to 10000
# taking their mean and appending it to list means.
for j in num:
    # Generating seed so that we can get same result
    # every time the loop is run...
    numpy.random.seed(2)
    x = [numpy.mean(
        numpy.random.randint(
            100, 6000, j)) for _i in range(500)]
    means.append(x)
k = 0
 
# plotting all the means in one figure
fig, ax = plt.subplots(2, 2, figsize =(10, 10))
for i in range(0, 2):
    for j in range(0, 2):
        # Histogram for each x stored in means
        ax[i, j].hist(means[k], 10, density = True)
        ax[i, j].set_title(label = num[k])
        k = k + 1
plt.show()
```
