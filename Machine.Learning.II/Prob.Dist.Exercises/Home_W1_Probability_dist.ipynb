Q1: Calculate the filled-up area under each normal/gaussian curve below:

Manually, we can find the area/probability value from Z-table.

```python
#z= 0.75

import scipy.stats as st
#when we look that up in the z-table by checking 0.7 under 5, we find 0.7733, which is the area to the left of z in above shape
print("P(Z<0.75) = ",st.norm.cdf(0.75))
p = st.norm.cdf(0.75)


print("So the probability that the z is greater than 0.75 is: ",p*100,"%")
```

```python
#z= -0.75

import scipy.stats as st
#when we look that up in the z-table by checking -0.7 under 5, we find 0.2266, which is the area to the righ of z in above shape
print("P(Z>-0.75) = ",st.norm.cdf(-0.75))
p = st.norm.cdf(-0.75)

#here we subtract the area from 1 to obtain the greater than area, since the total area under the curve is 1.
p = 1-p
print("1- P(Z>-0.75) = ", p*100)


print("So the probability that the z is greater than -0.75 is: ", p*100,"%")
```

```python
#P(-0.6< Z >1.2) 

import scipy.stats as st
#when we look that up in the z-table by checking -0.6 under 0, we find 0.2742
print("P(Z>-0.6) = ",st.norm.cdf(-0.6))

#when we look that up in the z-table by checking 1.2 under 0, we find 0.8849
print("P(Z>1.2) = ",st.norm.cdf(1.2))

p = (st.norm.cdf(1.2))-(st.norm.cdf(-0.6))
print("Area/Probability: ", p)

print("So the probability that the z between -0.6 and 1.2 is: ", p*100,"%")
```

```python
#P(-0.9< Z >1.6) 

import scipy.stats as st
#when we look that up in the z-table by checking -0.9 under 0, we find 0.1840
print("P(Z>-0.9) = ",st.norm.cdf(-0.9))

#when we look that up in the z-table by checking 1.6 under 0, we find 94520
print("P(Z>1.6) = ",st.norm.cdf(1.6))

p = (st.norm.cdf(1.6))-(st.norm.cdf(-0.9))
print("Area/Probability: ", p*100)


print("So the probability that the z between -0.9 and 1.6 is: ", p*100,"%")
```

# Confidence Interval

# Question No 1:
Q1: Construct a 95% confidence interval for the mean height of male Pakistanis. Forty-nine male Pakistanis are surveyed. The sample mean is 71 inches. The sample standard deviation is 2.8 inches. 

•	State the CI interval
•	Draw the graph
•	Calculate the error bound
•	What will happen to the level of confidence obtained if 2,000 male Pakistanis are surveyed instead of 49? Why?


```python
import math
from matplotlib import pyplot
import scipy.stats as st

mu=71
sd = 2.8
n=49
z=1.96 #95% interval z will be 1.96

#formula of CI
#Confidence Interval = x  +/-  t*(s/√n)

#Error of Bound
EBD = z*sd/math.sqrt(n)
print("Error of Bound (EBD)",EBD)

#confidence_interval = scipy.stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)
#confidence_interval = scipy.stats.t.interval(95, n-1, 175, margin_of_error)

#State the CI interval
lower = mu - EBD
upper = mu + EBD

f_s= sd/math.sqrt(n)
d = st.norm(mu, f_s)

#True mean of all Men height is likely to be between lower and upper 
#The "95%" says that 95% of experiments like we just did will include the true mean, but 5% won’t.
#So, there is a 1-in-2.8 chance (5%) that our Confidence Interval does NOT include the true mean.

print("CI Interval:",lower, upper)
print("The true mean of all Men height in Pakistan, is likely to be between",lower, "and", upper)

values = [value for value in range(65, 75)]
prob = [d.pdf(value) for value in values]
pyplot.plot(values, prob)
pyplot.axvline(lower, 0, 1)
pyplot.axvline(upper,0,1)
pyplot.show()
```

# Question No 2:
Q2: We do a survey to estimate the time needed to complete one person’s tax forms. We randomly survey 1000 people. The mean is 33.8 hours with sd of 6.5 hours. Assume normality
•	State the CI interval
•	Draw the graph
•	Calculate the error bound

```python
n= 1000
mu = 33.8
sd = 6.5
z=1.96


#Error of Bound
EBD = z*sd/math.sqrt(n)
print("Error of Bound (EBD)",EBD)


f_s= sd/math.sqrt(n)
d = st.norm(mu, f_s)

#State the CI interval
lower = mu - EBD
upper = mu + EBD


print("CI Interval:",lower, upper)
print("The time needed to compelte one person's tax form, is likely to be between",lower, "and", upper)

values = [value for value in range(25, 45)]
prob = [d.pdf(value) for value in values]
pyplot.plot(values, prob)
pyplot.axvline(lower, 0, 1)
pyplot.axvline(lower,0,1)
pyplot.show()
```

# Central Limit Theorem 

# Question No 1:
Q1: Model any problem from the agricultural domain in Python and demonstrate the central limit theorem (like dice roll and coin toss example).

The central limit theorem states that a sampling distribution of a sample statistic approaches the normal distribution as you take more samples, no matter the original distribution being sampled from.

If you’re a farmer and want to put in an order for pestiside, you’ll need to know how many kgs of pestiside your crop typically need.

Then you can take a different number of samples, all with the same size, and plot the pestidise data just to see how it looks like.

```python
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline 

num_i = np.asarray([2,5,15,50,100, 1000]); #sample

experiment = 'pesticide'  

trials = {'pesticide':10 } #sample space 10

nSamp=100000

k = trials[experiment] 

fig, fig_axes = plt.subplots(ncols=3, nrows=2, constrained_layout=True) 

print("Distribution of pest data from differnt sample size")

for i,N in enumerate(num_i):  
    y = np.random.randint(low=1,high=k+1,size=(N,nSamp)).sum(axis=0)  
    row = i//3;col=i%3;  
    bins=np.arange(start=min(y),stop=max(y)+2,step=1)  
    fig_axes[row,col].hist(y,bins=bins,density=True)  
    fig_axes[row,col].set_title('{} sample of {}'.format(N,experiment)) 
    
plt.show() 
```
